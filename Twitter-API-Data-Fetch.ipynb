#### Defining all the constants and global variables

```python
hashtags = ["AdaniGroup", "AdaniHindenburg", "HindenburgReport", "SEBI"]
```

```python
# CONSTANTS.
MAX_RESULTS = "100"
URL = f"https://api.twitter.com/2/tweets/search/recent?query={hashtags[3]}"
HEADERS = {"Authorization": "Bearer {}".format(BEARER_TOKEN)}
MAX_TWEETS = 200000

# variables that will change.
tweet_response = list()
tweet_count = 0
next_token = ''
current_url = URL
```

#### Loop to keep fetching the data untill the condition is True

```python
while tweet_count < MAX_TWEETS:
    print(tweet_count, next_token)
    response = requests.request("GET", current_url, headers=HEADERS)
    response = response.json()
    tweet_response.extend(response["data"])
    tweet_count += int(MAX_RESULTS)
    
    next_token = response["meta"].get("next_token", '')
    if(next_token == ''):
        break
    current_url = URL + "&max_results="+MAX_RESULTS
    current_url = current_url+"&next_token="+next_token
```

#### Another loop to seperate the tweet_id and tweet to be able to make dataframe

```python
tweet_ids = list()
tweets = list()

for tweet_res in tweet_response:
    tweet_ids.append(tweet_res["id"])
    twt = tweet_res["text"]
    tweets.append(twt)
```

#### Saving the dataframe into CSV

```python
df = pd.DataFrame({"Tweet ID": tweet_ids, "Tweet": tweets})
df.to_csv(f"{hashtags[3]}-02-03-2023.csv", index=False)
```

```python
import tweepy
import snscrape.modules.twitter as sntwitter
import numpy as np
```

```python
search_item = '#AdaniGroups'
since = "2023-01-25"
until = "2023-02-14"

# mysearch = f'{search_item} -filter:retweets since:{since} until:{until}'
```

```python
auth = tweepy.OAuthHandler(API_KEY, API_KEY_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)

myAPI = tweepy.API(auth, wait_on_rate_limit = True)
```

```python
def search(search_item, since, until, count):
    mysearch = f'{search_item} -filter:retweets since:{since} until:{until}'
    # collect tweet ids!!
    myscraper = sntwitter.TwitterSearchScraper(mysearch).get_items()
    print(myscraper)
    tweets_id = list()
    for i, tweet in enumerate(myscraper):
        if i>count:
            break
        print(tweet.id, tweet.content)
        tweets_id.append([tweet.id, tweet.date, tweet.content, tweet.username])

    # Status object from the Tweeter API
    mytweets = list()
    for myid in np.array(tweets_id)[:,0]:
        status = myAPI.get_status(myid, tweet_mode="extended")
        parsed_tweet = dict()
        # date and time of creation
        parsed_tweet['date'] = status.created_at
        parsed_tweet['id'] = status.user.id
        parsed_tweet['text'] = status.full_text #remove_url( status.full_text)
        mytweets.append(parsed_tweet)

    df_tweet = pd.DataFrame(mytweets)
    # just in case, remove duplicates
    df_tweet.drop_duplicates('text', keep='first', inplace=True)
    df_tweet.set_index('id', inplace=True)
    
    return(df_tweet)
```

```python
save = search(search_item, since, until, int(MAX_RESULTS))
```
